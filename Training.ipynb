```python
import zipfile
import os

zip_path = 'data.zip'
extract_path = '/kaggle/working/extracted_data'

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    for member in zip_ref.namelist():
        try:
            zip_ref.extract(member, extract_path)
        except Exception as e:
            print(f"فشل فك ضغط الملف: {member} بسبب {e}")

print("انتهت العملية!")
```

```python
def count_files(path):
    return len([f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f))])

train_img_path = '/kaggle/working/extracted_data/Foods/train/images'
train_lab_path = '/kaggle/working/extracted_data/Foods/train/labels'

if os.path.exists(train_img_path):
    print(f"عدد الصور: {count_files(train_img_path)}")
    print(f"عدد التسميات: {count_files(train_lab_path)}")
else:
    print("المسار غير صحيح، تأكد من مكان مجلد Foods")
```

```python
from ultralytics import YOLO

Model = YOLO('yolo11m-seg.pt')
results = Model.train(data='/kaggle/working/extracted_data/Foods/data_kaggle.yaml',
    epochs=140,
    patience=35,
    imgsz=640,
    batch=16,
    augment = True,
    copy_paste=0.4,
    mosaic=0.8,                  
    close_mosaic=20,
    mixup=0.1,
    translate=0.1,
    scale=0.5,
    cls = 1.65,
    device=[0,1])
```

```python
best_model = YOLO('/content/runs/segment/train3/weights/best.pt')
best_model.export(format='tflite', imgsz=640, int8=False)
```
